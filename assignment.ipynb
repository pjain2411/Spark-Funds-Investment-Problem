```python
# scrapping country names from wikipedia
import pprint, requests, bs4, numpy as np
url = "https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language"
req = requests.get(url)
# create a bs4 object
# To avoid warnings, provide "html5lib" explicitly
soup = bs4.BeautifulSoup(req.text, "html5lib")
divcountry = soup.select('.mw-parser-output')
countrybody = [link for r in divcountry for c in r.find_all('table') for tb in c.findNext('tbody') for rows in c.find_all('tr') for tds in rows.find_all("td") for link in tds.findNext("a",title=True)]
li=[str(result) for result in countrybody]
duplicatecountrynames=np.array(li)
countrynames=np.unique(duplicatecountrynames)
print(countrynames)
```

# Checkpoint 4 Sector Analysis

```python
#Extracting primary sector from master dataframe
master["primary_category"]=master["category_list"].str.split("|",n=1,expand=True)[0]
master.head()
```

```python
master.isna().sum()
#since the count of null values in primary_category is 579 need to impute these values
master["primary_category"].describe()

#imputing the null values to map in Others sector therfore replacing its value to category : All Markets
# master["primary_category"].fillna("All Markets",inplace=True)

# # Imputing with All Markets is not giving the correct model therefore dropping the nan values to get the accurate model --updated after model evaluation
master.dropna(subset=['primary_category'], inplace=True)
master.isna().sum()
```

```python
#Reading the mapping file
df_mapping = pd.read_csv("mapping.csv",encoding="ISO 8859-1")
df_mapping["category_list"]=df_mapping.category_list.str.encode('utf-8').str.decode('ascii','ignore')
df_mapping.head()
```

```python
#drop blank column
df_mapping.drop(["Blanks"],axis=1,inplace=True)
df_mapping.head()
```

```python
#drop Nan row
df_mapping.dropna(axis=0,inplace=True)
df_mapping.head()
```

```python
# creating dataframe for category sector mapping
df_sector=pd.melt(df_mapping,id_vars=['category_list'])
df_sector.head(20)
df_sector=df_sector[df_sector['value']==1]
df_sector.drop('value',axis=1,inplace=True)
df_sector=df_sector.rename(columns={'category_list':'primary_category','variable':'sector'})
df_sector
```

```python
# #merged data frame with each primary sector mapped to its main sector 
master=pd.merge(master,df_sector,how="inner",on="primary_category")
master.head(10)
```

```python
master=master.rename(columns={'primary_category':'primary_sector','sector':'main_sector'})
master.head()
```

```python
# Total amt of investment in each main sector
D1["raised_amount_usd"].sum()
D2["raised_amount_usd"].sum()
D3["raised_amount_usd"].sum()

```
